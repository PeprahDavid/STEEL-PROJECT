# DATA PREPROCESSING

```python
dataframe = pd.read_csv('CCT-TFS syyskuu Henry.csv', encoding = 'utf-8')
data=dataframe
data=data.drop(['Steel'], axis = 1)




data.describe()
#data
```

```python
x=data.drop(['Texp'], axis = 1)
y=data.Texp
x=x.astype(float)
y=y.astype(float)
```

```python

mc = MeanCenterer().fit(data)
centered_data=mc.transform(data)

scaler=preprocessing.MinMaxScaler()
data[data.columns.values] = scaler.fit_transform(centered_data)
```

```python
log_transformed = np.log(data)
log_transformed.replace([np.inf, -np.inf], np.nan, inplace=True)
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean',verbose=0)
imputer = imputer.fit(log_transformed)
int_data = imputer.transform(log_transformed)
int_data = pd.DataFrame(int_data)
#print(int_data)
int_data = int_data.set_axis(['C', 'Si', 'Mn', 'Cr', 'Mo', 'Ni', 'Al', 'Cu', 'ppmB', 'Pa', 'Rc',
       'lnRc', 'Texp'], axis=1, inplace=False)


int_data1=int_data.drop(['Texp','Pa','Rc','lnRc'], axis = 1)
```
