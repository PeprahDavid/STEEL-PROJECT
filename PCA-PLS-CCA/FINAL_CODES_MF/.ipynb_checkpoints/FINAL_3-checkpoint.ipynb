```python
#THE CCT DATA
dataframe = pd.read_csv('nthinteractiondata.csv', encoding = 'utf-8')
#dataframe
#print(dataframe.shape)
```

```python
#NORMALIZING THE DATA
dat=dataframe
scaler=MinMaxScaler()
dat[dat.columns.values] = scaler.fit_transform(dat)
transformed_data = dat.transform(func = lambda x : np.log(x))

#LOG TRANSFORM
df = transformed_data.fillna(0)
data = transformed_data
#data
```

```python
data = pd.DataFrame(dat, columns = ['C','Si','Mn','Cr', 'Mo','Ni','Al', 'Cu',
       'ppmB', 'Mn | Mn', 'C|Mn|Al', 'C|Al|Mn', 'Mn|Al|C', 'Al|C|Mn', 'Al|Mn|C',
       'Mn|C|Al', 'C | C', 'Ni | Cu', 'Cu | Ni', 'Cu | Mn', 'Mn | Cu',
       'Al|C|Mo', 'Mo|Al|C', 'Al|Mo|C', 'Mo|C|Al', 'C|Al|Mo', 'C | Mn',
       'Pa', 'Rc', 'lnRc','Texp'] )
```

```python
#DEPENDENT AND INDEPENDENT VARIABLES
#x=Dependent Variables
#y=Independent Variables

#data = pd.DataFrame.values
x=dat.drop(['Texp'], axis = 1)
y=dat.Texp
x=x.astype(float)
y=y.astype(float)
#x, y

#GETTING THE RESIDUALS
model_Simple = sm.OLS(y, x).fit()
x2 = sm.add_constant(x, prepend=True, has_constant='skip')
model = sm.OLS(y, x2).fit()
residuals=model.resid
#model.summary()
```

# DATA DIVISION

```python
train_ratio = 0.75
validation_ratio = 0.15
test_ratio = 0.10

X=x
y=y
sss = StratifiedShuffleSplit(n_splits=3, test_size=0.5,   random_state=42) 

StratifiedShuffleSplit(n_splits=10, test_size=None, train_size=None, random_state=None)
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                        test_size=1 - train_ratio)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, 
                                                test_size=test_ratio/(test_ratio + validation_ratio)) 
```

# RIDGE REGULARIZATION

```python
#ols_pred = ols_reg.predict(X_test)

ridge_df = pd.DataFrame({'variable': x.columns})
ridge_train_pred = {}
ridge_test_pred = {}

alphas = np.arange(0, 2, 0.1)

for n, alpha in enumerate(alphas):
    if (n<20):
        ridge_reg = Ridge(alpha=alpha)
        ridge_reg.fit(X_train, y_train)
        var_name = 'alpha ' + str(round(alpha,3))
        ridge_df[var_name] = ridge_reg.coef_
        # prediction
        #ridge_train_pred['alpha '+str(round(alpha,3))]=ridge_reg.predict(X_train)
        ridge_test_pred['alpha '+str(round(alpha,3))]=ridge_reg.predict(X_test)
    else:
        pass

#lasso_df
ridge_df =ridge_df.set_index('variable').T

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(ridge_df.C,'r', ridge_df.Si, 'g', ridge_df.Mn, 'b', ridge_df.Cr, 'c', ridge_df.Al, 'k')
ax.set_xlabel("alpha")

#ax.set_xticklabels(np.arange(-1, 10, 1))
ax.axhline(y=0, color='red', linestyle='--')
ax.set_xlabel("Lambda")
ax.set_ylabel("Beta Estimate")
ax.set_title("Ridge Regression Trace", fontsize=12)
ax.legend(labels=['Carbon','Si','Mn','Cr','Mo', 'NiMo', 'SiMo'])
fig.autofmt_xdate(rotation=45)
ax.grid(True)
#len(ols_pred)

```

```python
model = sm.OLS(y_train, X_train)
#model = sm.OLS(y, x)
results_fu = model.fit()
#print results_fu.summary()
results_fr = model.fit_regularized(L1_wt=0, alpha=0.00001, start_params=results_fu.params)
results_fr_fit = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)

final = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)

final.summary()
```

# LASSO REGULARIZATION

```python

lasso_df = pd.DataFrame({'variable': x.columns})
lasso_train_pred = {}
lasso_test_pred = {}

alphas = np.arange(0, 2, 0.001)
#alphas = np.arange(0.00001, 0.1, 0.00001)
#alphas = np.arange(0.01, 8.01, 0.4)

for n, alpha in enumerate(alphas):
    if (n<20):
        lasso_reg = Lasso(alpha=alpha)
        lasso_reg.fit(X_train, y_train)
        var_name = 'alpha ' + str(round(alpha,5))
        lasso_df[var_name] = lasso_reg.coef_
        # prediction
        #lasso_train_pred['alpha '+str(round(alpha,3))]=lasso_reg.predict(X_train)
        lasso_test_pred['alpha '+str(round(alpha,3))]=lasso_reg.predict(X_test)
    else:
        pass
        
#lasso_df
lasso_df =lasso_df.set_index('variable').T

fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(lasso_df.C,'r', lasso_df.Si, 'g', lasso_df.Mn, 'b', lasso_df.Cr, 'c', lasso_df.Al, 'k' )
ax.set_xlabel("alpha")

#ax.set_xticklabels(np.arange(-1, 10, 1))
ax.set_ylabel("Beta Estimate")
ax.set_title("Lasso Regression Trace", fontsize=12)
ax.legend(labels=['Carbon','Si','Mn','Cr','Mo', 'NiMo', 'SiMo'])
fig.autofmt_xdate(rotation=45)
ax.grid(True)

```

```python
model = sm.OLS(y_train, X_train)
#model = sm.OLS(y, x)

results_fu = model.fit()
#print results_fu.summary()
results_fr = model.fit_regularized(L1_wt=1, alpha=0.00001, start_params=results_fu.params)
results_fr_fit = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)

final = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)

#final.summary()
```

# ELASTIC_NET REGULARIZATION

```python
#print(x_train, x_val, x_test)
elastic = linear_model.ElasticNet(alpha=0)
elastic.fit(X_train, y_train)
elastic_df = pd.DataFrame({'variable': x.columns})
elastic_train_pred = {}
elastic_test_pred = {}
elastic_val_pred = {}

alphas = np.arange(0.00001, 0.1, 0.00001)
#alphas = np.arange(0.0003, 0.1, 0.00001)
#alphas = np.arange(0, 2, 0.001)


for n, alpha in enumerate(alphas):
    if (n<30):
        elastic_reg =linear_model.ElasticNet(alpha=alpha)
        elastic_reg.fit(X_train, y_train)
        var_name = 'alpha ' + str(round(alpha,5))
        elastic_df[var_name] = elastic.coef_
        # prediction
        #elastic_train_pred['alpha '+str(round(alpha,3))]=elastic_reg.predict(X_train)
        elastic_test_pred['alpha '+str(round(alpha,5))]=elastic_reg.predict(X_test)
        elastic_val_pred['alpha '+str(round(alpha,5))]=elastic_reg.predict(X_val)

    else:
        pass

elastic_df =elastic_df.set_index('variable').T


fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(elastic_df.C,'r', elastic_df.Si, 'g', elastic_df.Mn, 'b', elastic_df.Cr, 'c', elastic_df.Al, 'k' )
ax.set_xlabel("alpha")


#ax.set_xticklabels(np.arange(-1, 10, 1))
ax.set_ylabel("Beta Estimate")
ax.set_title("Elastic Net Regression Trace", fontsize=12)
ax.legend(labels=['Carbon','Si','Mn','Cr','Mo', 'NiMo', 'SiMo'])
fig.autofmt_xdate(rotation=45)
ax.grid(True)
```

```python


model = sm.OLS(y_train, X_train)
#model = sm.OLS(y, x)

results_fu = model.fit()
#print results_fu.summary()

results_fr=model.fit_regularized(method='elastic_net', L1_wt=0.34, alpha=0.00001,
                                 start_params=results_fu.params, refit=False)

#results_fr = model.fit_regularized(method='elastic_net', L1_wt=0.5, alpha=0.00001, start_params=results_fu.params)
results_fr_fit = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)
#GLM.fit_regularized(method='elastic_net', alpha=0.00001, start_params=results_fu.params, refit=False)

final = sm.regression.linear_model.OLSResults(model, results_fr.params, model.normalized_cov_params)

final.summary()
```
